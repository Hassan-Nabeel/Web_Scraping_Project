```python
import chromedriver_binary
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm

import numpy as np
import pandas as pd
```

```python
browser = webdriver.Chrome()

browser.get('https://www.youtube.com/c/GeeksforGeeksVideos/videos')
```

```python
soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
data = []

for sp in soup.find_all('ytd-rich-item-renderer'):
    
    title = sp.find('a',class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').text
    video_link = 'https://www.youtube.com' + sp.find('a',class_='yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media').get('href')
    
    try:
        f = sp.find_all('span',class_='inline-metadata-item style-scope ytd-video-meta-block')
        views = f[0].text
        date = f[1].text
    except:
        views = np.nan     # None
        date = np.nan      # None
        
    try:
        thumbnail_link = sp.find('img').get('src').split('?')[0]
    except:
        thumbnail_link = np.nan   # None
        
    data.append([title, video_link, views, date, thumbnail_link])
    
```

```python
df = pd.DataFrame(data, columns = ['title', 'video_link','views','date','thumbnail_link'])
```
