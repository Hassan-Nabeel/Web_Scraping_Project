## Scraping content / text form Wikipedia

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Martin_Luther_King_Jr.'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
heading = soup.find('h1').text

heading
```

```python
corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()
    
```

```python
print(corpus)
```

```python
for i in range(2,467):
    corpus = corpus.replace('['+str(i)+']', '')
    
```

```python
print(corpus)
```

```python
f = open(heading + '.txt', 'w', encoding='utf-8')
f.write(corpus)
f.close()
```

## Creating a new link and scraping data form it

```python
title = input('Enter the name: ').replace(' ','+')
url = 'https://www.google.com/search?q=' + title + '+wikipedia'
# print(url)

res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')

for sp in soup.find_all('div'):
#     print(sp.text)

    try:
        link = sp.find('a').get('href')
        if ('en.wikipedia.org' in link):
            break
    except:
        pass
    
link = link[7:].split('&')[0]
print(link)
```

```python
soup.text
```

```python
res = requests.get(link)
soup = BeautifulSoup(res.text, 'html.parser')

corpus = ''

for p in soup.find_all('p'):
    corpus += p.text
    corpus += '\n'
    
corpus = corpus.strip()

for i in range(2,467):
    corpus = corpus.replace('['+str(i)+']', '')
    
```

```python
print(corpus)
```
