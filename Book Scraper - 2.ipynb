### Approach - 2

```python
data = []

for link in tqdm(df['book_link']):
    
    res =  requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
#     print(soup.text)

    typ           = soup.find('ul', class_ = 'breadcrumb').find_all('a')[2].text
    
    varible       = soup.find('table', class_ = 'table table-striped').find_all('td')
    
    UPC           = varible[0].text
    product_type  = varible[1].text
    price_excl    = varible[2].text[2:]
    price_incl    = varible[3].text[2:]
    tax           = varible[4].text[2:]
    availability  = varible[5].text
    reviews       = varible[6].text
    
    data.append([typ, price_excl, price_incl, tax, availability, UPC, reviews])
    
```

```python
df2 = pd.DataFrame(data, columns = ['category', 'price_ex_tax', 'price_in_tax', 'tax', 'quantity', 'upc', 'review'])
```

```python
df2.head()
```

```python
df2.to_csv('data.csv', index=True)
```

## Data Combining of Book Scraper-1 and Book Scraper-2

```python
df_1 = pd.read_csv('books.csv')
df_2 = pd.read_csv('data.csv')
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
# creating new dataframe

df = pd.DataFrame()

df['title']         = df_1['title']
df['upc']           = df_2['upc']
df['category']      = df_2['category']
df['price_ex_tax']   = df_2['price_ex_tax']
df['price_in_tax']   = df_2['price_in_tax']
df['tax']           = df_2['tax']
df['rating']        = df_1['rating']
df['review']       = df_2['review']
df['quantity']      = df_2['quantity']
df['stock']         = df_1['stock']

df['book_link']     = df_1['book_link']
df['image_link']      = df_1['image_link']
```

```python
df.to_csv('file.csv', index=False)
```
